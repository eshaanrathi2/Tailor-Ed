```python
import json
from pathlib import Path
```

```python

file_path='../data/secondary_data.json'
json_data = json.loads(Path(file_path).read_text())
```

```python
for data in json_data:
    print(data, json_data[data])
```

```python
MATH_docs = []
```

```python
for key in json_data:
    if "MATH" in key:
        MATH_docs.append(key)
```

```python
print(len(MATH_docs), MATH_docs)
```

```python
# Filter the data for keys with the value containing "MATH"
filtered_data = {k: v for k, v in json_data.items() if isinstance(v, str) and "MATH" in v}

# Print the filtered data
print(filtered_data)
```
